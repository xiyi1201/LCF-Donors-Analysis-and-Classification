```python
event = pd.read_csv('merge_file.csv',low_memory=False)
event = event.loc[:519792]
event = event.dropna(subset=['Constituent ID'])
event['Constituent ID'] = event['Constituent ID'].astype(int)
```

```python

event['Gift Amount'] = event['Gift Amount'].replace('[\$,]', '', regex=True).astype(float)

# Check the new data type of the column
print(event.dtypes)

event['Appeal Split Amount'] = event['Appeal Split Amount'].replace('[\$,]', '', regex=True).astype(float)
donor['Appeal Split Amount'] = donor['Appeal Split Amount'].replace('[\$,]', '', regex=True).astype(float)


# Check the new data type of the column
print(event.dtypes)
```

```python


# Identify common columns with the same name and data type
common_columns = [col for col in donor.columns if col in event.columns and donor[col].dtype == event[col].dtype]

# Merge the dataframes on these common columns
result = pd.merge(donor, event, on=common_columns, how='inner')

# Display the result
result
```
