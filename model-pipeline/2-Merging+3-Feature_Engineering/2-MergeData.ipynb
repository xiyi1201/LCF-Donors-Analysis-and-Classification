```python
donor_file = "giftandzipdataindexed.csv"
gift_file = "merge_file.csv"
```

```python
import pandas as pd
# read donor data 
donor = pd.read_csv(donor_file, low_memory=False)
donor.head()
```

```python
# read gift data 
event = pd.read_csv(gift_file, low_memory=False)
# data cleaning
event = event.loc[:519792]
event = event.dropna(subset=['Constituent ID'])
event['Constituent ID'] = event['Constituent ID'].astype(int)
event['Gift Amount'] = event['Gift Amount'].replace('[\$,]', '', regex=True).astype(float)
event['Appeal Split Amount'] = event['Appeal Split Amount'].replace('[\$,]', '', regex=True).astype(float)
donor['Appeal Split Amount'] = donor['Appeal Split Amount'].replace('[\$,]', '', regex=True).astype(float)
```

```python
# merge two datasets
result = pd.merge(donor, event, on='Index', how='inner')
result.drop('Constituent ID_x',axis=1)
```
